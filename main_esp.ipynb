## 6. Escribe una funci贸n que devuelva el factorial de un n煤mero.

```python
def factorial(n):
    if n < 0:
        raise ValueError("El factorial no est谩 definido para n煤meros negativos.")
    elif n == 0:
        return 1
    else:
        resultado = 1
        for i in range(1, n + 1):
            resultado *= i
        return resultado
```

## 8. Escribe una funci贸n que devuelva la moda de una lista, es decir: el elemento que aparece m谩s veces.

`NOTE: No se debe usar count... `

```python
def mode_counter(arr):
    frequency = {} 
    for item in arr:
        if item in frequency:
            frequency[item] += 1
        else:
            frequency[item] = 1

    max_frequency = max(frequency.values())
    mode = None
    for key, value in frequency.items():
        if value == max_frequency:
            mode = key
            break

    return mode
```

## 9. Escribe una funci贸n que calcule la desviaci贸n est谩ndar de una lista.
`NOTE: no utilices librer铆as ni ninguna funci贸n ya construida. `

```python
def st_dev(list_sd):
    mean = sum(list_sd) / len(list_sd)

    
    sum_squared_diff = sum((x - mean) ** 2 for x in list_sd)


    variance = sum_squared_diff / len(list_sd)

 
    std_dev = variance ** 0.5

    return std_dev
```

## 10. Escribe una funci贸n para verificar si una cadena es un pangrama, es decir: si contiene todas las letras del alfabeto al menos una vez. Ten en cuenta que las cadenas pueden contener caracteres que no son letras.

```python
def pangram(string):
    
    alphabet_set = set('abcdefghijklmnopqrstuvwxyz')

    cleaned_string = ''.join(filter(str.isalpha, string.lower()))

    string_set = set(cleaned_string)

    return alphabet_set.issubset(string_set)
```

## 11. Escribe una funci贸n que reciba una cadena de palabras separadas por comas y devuelva una cadena de palabras separadas por comas ordenadas alfab茅ticamente.

`NOTA: Puedes usar sorted pero no split y definitivamente no join! お`

```python
def sort_alpha(string):
    words = ['']
    for char in string:
        if char == ',':
            words.append('')
        else:
            words[-1] += char
    
    sorted_words = sorted(words)
    
    for word in sorted_words:
        sorted_string += word + ','
    
    return sorted_string[:-1]
```

```python
# Para testear tu funci贸n 
test_alpha(sort_alpha)
```

## 12. Escribe una funci贸n para verificar si una contrase帽a dada es fuerte (al menos 8 caracteres, al menos una min煤scula, al menos una may煤scula, al menos un n煤mero y al menos un car谩cter especial). Deber铆a devolver True si es fuerte y False si no lo es.

```python
def check_pass(password):
    if len(password) < 8:
        return False

    has_lower = any(char.islower() for char in password)
    has_upper = any(char.isupper() for char in password)
    has_digit = any(char.isdigit() for char in password)
    has_special = any(char in '!@#$%^&*()-_=+[]{}|;:,.<>?/~' for char in password)

    return has_lower and has_upper and has_digit and has_special
```

```python
# Para testear tu funci贸n
test_pass(check_pass)
```
